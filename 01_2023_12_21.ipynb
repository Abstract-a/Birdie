**Excercise 1.A** Write a function that takes two positive integers `n` and `m` as arguments, and returns a matrix with `n` rows and `m` columns whose values are random numbers in the range [0,1) - you can use the method `random.random()`

```python
import random

def create_matrix(n, m):
  M = [[random.random() for _ in range(m)] for _ in range(n)]
  return M
```

**Excercise 1.B** Write a function that takes as input a matrix `M` and a scalar value `v`, and returns a matrix that corresponds to the product $v \cdot M$.

```python
def multiply(M, v):
  n = len(M)
  m = len(M[0])
  Q = [[v*M[i][j] for j in range(m)] for i in range(n)]
  return Q
```

**Excercise 2.A** Write a function that takes as input a matrix M and writes the matrix in a file, such that each row of the matrix goes on a different row of the file.

```python
def save_matrix(M):
  myfile = open('matrix.txt', 'w')
  for i in range(len(M)):
    myfile.write(str(M[i])+"\n")
  myfile.close()
```
