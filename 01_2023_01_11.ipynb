**Excercise 1.A** Write a function that prompts the user to enter two positive integers. The program should warn the user if the input is invalid (i.e., the input is not a positive integer) and keep asking until a valid input is entered. Next, the program prints the two numbers. Try to avoid repeating code unnecessarily.

```python
def enter_integer():
  choice = ""
  while not choice.isdigit():
    choice = input("Enter a positive integer: ")
    if not choice.isdigit():
      print("The input is invalid.")
  return int(choice)

def main():
  n = enter_integer()
  m = enter_integer()
  print(f"The entered numbers are {n} and {m}.")

main()
```

**Excercise 1.B** Write a function that prompts the user to enter a matrix of size $n \times m$, where $n$ and $m$ are the first and the last integers entered before. Next, the program saves the matrix into a file. Note: the first two lines of the file should contain the number of rows and columns of the matrix.

```python
def enter_matrix(n, m):
  matrix = [[0 for _ in range(m)] for _ in range(n)]
  for i in range(n):
    for j in range(m):
      matrix[i][j] = input(f"Enter the element at position {i},{j}: ")
  return matrix

def save_matrix(matrix):
  file = open("matrix.txt", "w")
  n = len(matrix)
  m = len(matrix[0])
  file.write(str(n)+"\n")
  file.write(str(m)+"\n")
  for i in range(n):
    for j in range(m):
      file.write(matrix[i][j]+"\n")
  file.close()

def main():
  n = enter_integer()
  m = enter_integer()
  matrix = enter_matrix(n, m)
  save_matrix(matrix)

main()
```

**Excercise 2.A** Write a function that, given a matrix as input, computes its transpose. (Recall that the transpose of a matrix is an operator which flips a matrix over its diagonal; that is, it switches the row and column indices.)

```python
def transpose(matrix):
  n = len(matrix)
  m = len(matrix[0])
  return [[matrix[j][i] for j in range(n)] for i in range(m)]
```

**Excercise 2.B** Write a program that reads the matrix saved into the previous file (exercise 1.B), computes its transpose, and finally prints the transposed matrix.

```python
def read_matrix():
  file = open("matrix.txt", "r")
  n = int(file.readline())
  m = int(file.readline())
  matrix = [[0 for _ in range(m)] for _ in range(n)]
  for i in range(n):
    for j in range(m):
      matrix[i][j] = file.readline().rstrip()
  file.close()
  return matrix

def main():
  matrix = read_matrix()
  print(matrix)
  print(transpose(matrix))

main()
```
