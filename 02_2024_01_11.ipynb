**Exercise 1.A** Write a function that takes two lists of numbers and returns a new list that contains only the common elements between the two lists, without duplicates.

```python
def find_common(listA, listB):
  common = []
  for a in listA:
    if a in listB and a not in common:
      common.append(a)
  return common
```

**Exercise 1.B** Write a program that prompts the user to enter two lists of numbers. The two lists can have different lengths. The user should **not** specify in advance the length of each list. The program should warn the user if the entered values are not valid. Avoid duplicated code, if possible. Finally, use the previous function to find the common values in the two lists entered by the user.

```python
def enter_numbers():
  print("Enter a list of numbers, end with 'OK'.")
  values = []
  next_value=""
  while next_value != "OK":
    next_value = input("Enter the next value: ")
    if next_value != "OK":
      try:
        next_value_num = float(next_value)
        values.append(next_value_num)
      except ValueError:
        print("Please enter a valid value.")
  return values

def main():
  listA = enter_numbers()
  listB = enter_numbers()
  common = find_common(listA, listB)
  print("Common values: ",common)

main()
```

**Exercise 2.A** Write a function that takes a non-negative integer and returns the sum of its digits using recursion.

```python
def sum_of_digits(n):
  value = None
  if n < 10:
    value = n
  else:
    value = n % 10 + sum_of_digits(n // 10)
  return value
```

**Exercise 2.B** Write a program that generates 10 integers in the range $[0,100]$ and, for each of them, it uses the previous function to print the sum of its digits.

```python
import random

K = 10
MIN = 0
MAX = 100

def generate_numbers():
  return [random.randint(MIN, MAX) for _ in range(K)]

def main():
  numbers = generate_numbers()
  for n in numbers:
    print(f"Value {n}, sum of its digits: {sum_of_digits(n)}")

main()
```
